## WaP find the factorial

```python
a = int(input("Please enter a number"))
factorial = 1
for i in range(a+1):
    if (i==0):
        continue
    factorial = i * factorial
print(factorial)
```

```python
# Better Code

a = int(input("Please enter a number"))
factorial = 1

for i in range (1, a+1):
    factorial *=i
print(factorial)
```
