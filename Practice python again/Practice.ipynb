# Loops

### Print numbers from 1 to 10 using a for loop.

```python
for a in range(1, 11):
    print (a)
```

### Print only even numbers between 1 and 20 using a while loop.

```python
a = 1

while (a <= 20):
    if (a%2 == 0):
        print(a)
    a += 1
```

### Write a program to find the sum of numbers from 1 to n (where n is input from the user).
ðŸ‘‰ Example: If n = 5, output should be 15 because 1+2+3+4+5 = 15.

```python
n = int(input("Please enter a number"))

sum_total = 0
for i in range (1, n+1):
    sum_total += i
print(sum_total)
```

### Write a program that prints the multiplication table of a number (take input from user).

```python
n = int(input("Please enter a number"))

proDuct = 1

for i in range (1, 11):
    proDuct = n*i
    print(f"{n}x{i}={proDuct}")
    
```

### Q6. Reverse a String using Loop

Input: hello

Output: olleh

```python
a = "Hello"

for i in range(len(a)-1, -1, -1):
    print(a[i],end="")
```

```python
a = 'Hello'
rev = ''

for i in range(len(a)):
    rev = a[i] + rev
print (rev)
```

#### Step by Step

#### input("Enter numbers separated by space: ")

Asks the user for input.

Example: user types

10 5 8 12 3


This is a string: "10 5 8 12 3"

#### .split()

Splits the string into parts at spaces.

"10 5 8 12 3".split() â†’ ["10", "5", "8", "12", "3"]

Now itâ€™s a list of strings.

#### map(int, ["10", "5", "8", "12", "3"])

map(int, ...) applies the int function to each string.

Converts each "10" â†’ 10 (integer).

Result: map object that contains integers [10, 5, 8, 12, 3].

#### list(...)

Converts the map object into a real Python list.

Final result:

user_input = [10, 5, 8, 12, 3]

```python
user_input = list(map(int,input("Enter the element seprated by space: ").split()))

print (user_input)

max_value = user_input[0]
min_value = user_input[0]

for i in user_input:
    if i < min_value:
        min_value = i
    if i > max_value:
        max_value = i
print(f'Max value is {max_value} & Min Value is {min_value}')
        
```
