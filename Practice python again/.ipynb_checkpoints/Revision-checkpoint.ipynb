## Problem 1 : Factorial Using While Loop

```python
user_num = int(input('Please enter a number: '))

factorial = 1
x = 1

while (x <= user_num):
    factorial *= x
    x += 1

print(f"The factorial of {user_num} is {factorial}.")
```

## Factorial Using for Loop

```python
user_num = int(input("Please enter a number : "))

x = 1

for i in range (1, user_num+1):
    x = x*i
print(x)
    
```

## Problem 2: Removing the vowels and the special character form the name

```python
name = input("Please enter your name")

new_name = ""

for i in name:
    if i in 'aeiou' or i in 'AEIOU':
        continue
    if ord(i) >= 32 and ord(i)<=64:
        continue
    new_name = new_name+i
print(new_name)
```

```python
print(chr(34))
print(ord('"'))
```

What Could Be Improved form the previous code:

ASCII range 32–64 removes too many characters — it includes digits (0–9), which maybe you don't want to exclude unless specified.

It doesn’t handle special characters beyond 64 like [, ], {, }, ~, etc.

If the goal is only keeping alphabetic consonants, Python has a cleaner way.

```python
name = input("Please enter your name")

new_name = ""

for i in name:
    if i.lower() in 'aeiou':
        continue
    if not i.isalpha():
        continue
    new_name +=i
print(new_name)
```

## Problem : 3  Check Even or Odd

```python
num = int(input('Give me a number'))

if (num % 2 == 0):
    print("It is even")
else:
    print("It is odd")
```

## Problem 4 : String Multiplication + Type Casting

```python
name = str(input("Please enter your name"))
age = int(input("Please enter your age"))
centurion = 100
to_be = centurion - age

print(f'Hi, My name is {name} and i am {age} year old and I am {to_be} year away form being {centurion}')
```

```python
name = str(input("Please enter your name"))
age = int(input("Please enter your age"))
centurion = 100
if age < centurion:
    to_be = centurion - age
    print(f'Hi, My name is {name} and i am {age} year old and I am {to_be} year away form being {centurion}')
elif age == 100:
    print(f'Hi, My name is {name} and i am {age} year old and I am already {centurion}')
else:
    is_more = age - centurion
    print(f'Hi, My name is {name} and i am {age} year old and I am {is_more} more than {centurion}')
    
```

## Problem 5 : Pattern Printing – Descending Rows

1 1 1 1

3 3

4


```python
n = int (input("please enter a number"))

for i in range(1, n+1):
    for j in range (n):
        print(i,end = "")
    n -=1
    print()
```

### Imporved

```python
n = int (input("please enter a number"))

for i in range(1, n+1):
    for j in range (n-i+1):
        print(i,end = "")
    print()
```

#### this "try" and "except" helps the code not to crash

```python
my_dict = {}

def smart_casting_values(value):
    try:
        return int(value)
    except ValueError:
        try:
            return float(value)
        except ValueError:
            return value
    
while True:
    key = input("Please enter a key or type done to finish entering key")
    if key.lower() == 'done':
        break
    value = input (f"Please enter {key}")
    my_dict[key] = smart_casting_values(value)
        
       
print(my_dict)
```

## Problem 8: Factorial Using a List and Loop

```python
import sys
try:
    num = int(input("Please enter a number"))
except ValueError:
    print("This is not a number you dumb ass")
    sys.exit()
    
my_list = []

for i in range(1,num+1):
    my_list.append(i)
print(my_list)   
fac = 1

for i in my_list:
    fac *=i
print(fac)
```

## Problem 9: Print a Triangle of Repeated Numbers

```python
num = int(input("Please enter a number"))

for i in range (1,num+1):
    for j in range(i):
        print(i, end="")
    print()
```
