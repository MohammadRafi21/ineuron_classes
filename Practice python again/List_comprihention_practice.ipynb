# list comprehension

```python
# this is a normal code that we do
# in this code we are finding the square of a number

squares = []

for i in range(1,6):
    squares.append(i*i)

print(squares)
```

```python
# we can write the above code in 1 line using list comprehension

squares = [i*i for i in range (1,6)]
print(squares)
```

```python
# example 2

# I am finding the even numbers form the range 1 to 10

even = [i for i in range(1,11) if i%2 == 0]
print(even)
```

```python
# example 3
# findig the fruits name with a in it

fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
newlist = [f for f in fruits if 'a'in f]
print(newlist)
```

3Ô∏è‚É£ Replace Even Numbers with ‚Äúeven‚Äù

Given:

Make a new list ‚Üí ['odd', 'even', 'odd', 'even', ...].
üß† Hint: use if and else inside the expression part (before the for).

```python
nums = [1, 2, 3, 4, 5, 6]

newList = ['even' if i%2==0 else 'odd' for i in nums]
print(newList)
```

```python
nums = [1,2,3,4,5,6]
newList = []
for i in nums:
    if i%2 == 0:
        newList.append('even')
        
    else:
        newList.append('odd')
print(newList)
```

üß© Question 5: Dictionary from Two Lists

You‚Äôre given:

keys = ['name', 'age', 'city']
values = ['Rafi', 24, 'Melbourne']


Create a dictionary that looks like this:

{'name': 'Rafi', 'age': 24, 'city': 'Melbourne'}

```python
keys = ['name', 'age', 'city']
values = ['Rafi', 24, 'Melbourne']

newDict = {keys[i]:values[i] for i in range(len(keys))}
print(newDict)
# x = zip(keys,values)
```

```python
keys = ['name', 'age', 'city']
values = ['Rafi', 24, 'Melbourne']
newDict = {key : value for key,value in zip(keys,values)}
print(newDict)
```

üß© Problem 6 ‚Äî Multiplication Table (Nested Comprehension)

You need to create this:

[[1, 2, 3, 4, 5],

 [2, 4, 6, 8, 10],

 [3, 6, 9, 12, 15],

 [4, 8, 12, 16, 20],

 [5, 10, 15, 20, 25]]
 


Each inner list is a row for one number‚Äôs multiplication table.

```python
multTable = [[i*j for j in range(1,6)] for i in range(1,6)]
print(multTable)
```

üß© Problem 8 ‚Äî Unique Squares

You‚Äôre given:

nums = [1, 2, 2, 3, 3, 4]


You need to create a sorted list of unique squares ‚Üí
‚úÖ Output should look like:

[1, 4, 9, 16]

```python
nums = [1, 2, 2, 3, 3, 4]
newNums = [i*i for i in nums]
newNums1 = set(newNums)
newNums = list(newNums1)
newNums.sort()
print(newNums)
```
