```python
# Wap to take input form user his age name and marks and show output as

# 1. name in opposite direction (reversed)

# 2. age as how many years left in turning 100

# 3. marks as how much are they out of 100
```

```python
name = input("Please enter your name")
age = int(input("Please enter your age"))
marks = float(input("Please enter your marks"))
n = name[::-1]
age = 100-age
marks = (marks/100)*100
print("the name in reverse is ", n, "\n", "The years left to turn 100 is ", age, "\n", "The marks out of 100 is ", marks,"%")
```

# Print Statement

### Format specifier
### %d for int, %s for str, %f is for float

```python
name = 'Rafi'
age = 45
print("name is %s and age is %d"%(name,age))
```

```python
decimal = 35.163514
```

```python
print("Number is %d"%(decimal))
```

```python
print("Number is %f"%(decimal))
```

```python
print("Number is %s"%(decimal))
```

```python
print("Number is %0.2f"%(decimal))
```

```python
print("Number is %0.3f"%(decimal))
```

# .format Argument

```python
print("My name is {} and age is {}".format(name,age))
```

```python
name = input("Please enter your name")
age = int(input("Please enter your age"))
marks = float(input("Please enter your marks"))
n = name[::-1]
age = 100-age
marks = marks/100
print("the name in reverse is {} \n The years left to turn 100 is {} \n The marks out of 100 is {}".format(n,age,marks))
```

```python
print("My name is {0} and age is {1}".format(name,age))
```

```python
print("My name is {1} and age is {1}".format(name,age)) # here we can see they are taking the the position of a set if its given
```

```python
print("My name is {} and age is {1}".format(name,age)) # either all empty or all full
```

```python
print("My name is {name} and age is {age}".format(name=name,age=age))
```

```python
print("My name is {a} and age is {b}".format(a=name,b=age))
```
