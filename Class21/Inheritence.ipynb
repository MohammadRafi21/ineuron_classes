## Aggregation

```python
class PriceClass: # custom data type
    def __init__(self, price, curr):
        self.price = price
        self.currency = curr

    def getPrice(self):
        print(self.currency + " "+ str(self.price))
```

```python
class Product:
    def __init__(self, name, brand, priceVariable, l1):
        self.name = name
        self.brand = brand
        self.priceVariable = priceVariable
        self.l1 = l1

    def getDetails(self): 
        print(self.name)
        print(self.brand)
        self.priceVariable.getPrice()
        self.l1.append(5)
        print(self.l1)
```

```python
productPrice = PriceClass(20000,"AUD")
```

```python
print(productPrice.price)
print(productPrice.currency)
```

```python
iPhone = Product("iphone", "apple", productPrice, [1,2,3])
```

```python
print(iPhone.name)
print(iPhone.brand)
print(iPhone.priceVariable)
```

```python
print(iPhone.priceVariable.price)
```

```python
print(iPhone.priceVariable.currency)
```

```python
print(iPhone.l1)
```

```python
iPhone.getDetails()
```

 __we can see that the product class "has a" price in it__

 __this is called aggregation__

## super() method

__It is used to access properties of parent class__
__it also just work inside class__

```python
class iNeuron:

    def __init__(self):
        print("I am in iNeuron")

    def login(self):
        
        print("logged in")
```

```python
class Mentor(iNeuron): # here the class mentor is inheriting 
    # the class "iNeuron" 
    def __init__(self,speciality,name):
        print("I am in Mentor")
        self.speciality = speciality
        
    def login(self):
         super().login() # by this we can use the parents "login" method
         print("This is mentor login")
```

here parent have "login" and child have a "login" so super is used to show that I am calling the parents method

### method Overloading (polymorhism)

__same name method different input different output__

```python
class Shape:

    
    def area(self,radius):
        return 3.14 * radius**2
    
    def area(self, l, b):
        return l*b
```

```python
s= Shape()
```

```python
s.area(5)
```

```python
class Shape:
    
    def area(self, side = 3, breadth=0): # same name method  
        # different input different output
        if (breadth == 0):
            return 3.14 * side**2
        
        else:
            return side*breadth
        
```

```python
s= Shape()
```

```python
s.area(5)
```

```python
s.area(5,3)
```

```python
from abc import ABC,abstractmethod

class iNeuron(ABC):

    def __init__(self,name):
        self.name = name

    def login(self):
        print("logged in")

    
    def connect_to_DB(self,password):
         print("connect to DB")

    def access_recording(self):
         print("accessing recording")

    @abstractmethod
    def authorization(self):
        pass


    def take_test(self):
        print("taking test")
```

```python
class Student(iNeuron):
    def authorization(self):
        print("first we will do student authorization")
```

```python
class Mentor(iNeuron):
    pass
```

```python
s = Student("Rafi")
```

```python
s.authorization()
```

```python
m = Mentor("rafi")
```

__✅ If a parent class defines an abstract method, then the child class must implement (override) that method, otherwise the child class cannot be instantiated (used to create an object).__

__✅ If the parent has an abstract method, the child must write that method again. If not, we cannot create an object of the child class.__
