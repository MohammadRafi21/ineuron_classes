## Aggregation

```python
class PriceClass: # custom data type
    def __init__(self, price, curr):
        self.price = price
        self.currency = curr

    def getPrice(self):
        print(self.currency + " "+ str(self.price))
```

```python
class Product:
    def __init__(self, name, brand, priceVariable, l1):
        self.name = name
        self.brand = brand
        self.priceVariable = priceVariable
        self.l1 = l1

    def getDetails(self): 
        print(self.name)
        print(self.brand)
        self.priceVariable.getPrice()
        self.l1.append(5)
        print(self.l1)
```

```python
productPrice = PriceClass(20000,"AUD")
```

```python
print(productPrice.price)
print(productPrice.currency)
```

```python
iPhone = Product("iphone", "apple", productPrice, [1,2,3])
```

```python
print(iPhone.name)
print(iPhone.brand)
print(iPhone.priceVariable)
```

```python
print(iPhone.priceVariable.price)
```

```python
print(iPhone.priceVariable.currency)
```

```python
print(iPhone.l1)
```

```python
iPhone.getDetails()
```

 __we can see that the product class "has a" price in it__

 __this is called aggregation__

## Inheritance

```python
 class iNeuron:


    def login(self):
        print("logged in")
    def logout(self):
        print("Logged out")
```

```python
class Mentor(iNeuron): # here the class mentor is inheriting 
    # the class "iNeuron"
    pass
```

```python
class Student(iNeuron):# here the class Student is inheriting 
    # the class "iNeuron"
    pass
```

```python
m1 = Mentor()
```

```python
m1.login() # here the Mentor class has the properties of the class iNeuron
```

## Inheritance with constructor

```python
class Mentor(iNeuron): # here the class mentor is inheriting 
    # the class "iNeuron"
    pass
```

```python
class Student(iNeuron):# here the class Student is inheriting 
    # the class "iNeuron"
    pass
```

```python
m1 = Mentor("Rafi")
```

```python
m1.login()
```
