## Aggregation

```python
class PriceClass: # custom data type
    def __init__(self, price, curr):
        self.price = price
        self.currency = curr

    def getPrice(self):
        print(self.currency + " "+ str(self.price))
```

```python
class Product:
    def __init__(self, name, brand, priceVariable, l1):
        self.name = name
        self.brand = brand
        self.priceVariable = priceVariable
        self.l1 = l1

    def getDetails(self): 
        print(self.name)
        print(self.brand)
        self.priceVariable.getPrice()
        self.l1.append(5)
        print(self.l1)
```

```python
productPrice = PriceClass(20000,"AUD")
```

```python
print(productPrice.price)
print(productPrice.currency)
```

```python
iPhone = Product("iphone", "apple", productPrice, [1,2,3])
```

```python
print(iPhone.name)
print(iPhone.brand)
print(iPhone.priceVariable)
```

```python
print(iPhone.priceVariable.price)
```

```python
print(iPhone.priceVariable.currency)
```

```python
print(iPhone.l1)
```

```python
iPhone.getDetails()
```

 __we can see that the product class "has a" price in it__

 __this is called aggregation__

In inheritance, a child class can access only the public (and protected) methods of the parent, not its private variables directly__

## super() method

__It is used to access properties of parent class__
__it also just work inside class__

```python
class iNeuron:

    def __init__(self):
        print("I am in iNeuron")

    def login(self):
        
        print("logged in")
```

```python
class Mentor(iNeuron): # here the class mentor is inheriting 
    # the class "iNeuron" 
    def __init__(self,speciality,name):
        print("I am in Mentor")
        self.speciality = speciality
        
    def login(self):
         super().login() # by this we can use the parents "login" method
         print("This is mentor login")
```

here parent have "login" and child have a "login" so super is used to show that I am calling the parents method

```python
m1 = Mentor("DS","Rafi")
```

```python
m1.login()
```

```python
class iNeuron:

    def __init__(self,name):
        print("I am in iNeuron")
        self.name = name
        self.__income = "200"

    def login(self):
        print("logged in")
    def logout(self):
        print("Logged out")
    def getIncome(self):
        print(self.__income)
    def getName(self):
        print(self.name)
```

```python
class Mentor(iNeuron): # here the class mentor is inheriting 
    # the class "iNeuron" 
    def __init__(self,speciality,name):
        print("I am in Mentor")
        super().__init__(name)
        self.speciality = speciality
        
    def login(self):
         print("This is mentor login")

    def getDetail(self):
      print(self.name)
      print(self.__income)
```

```python
m1 = Mentor("DS","Rafi")
```
