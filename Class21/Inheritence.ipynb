```python
class Price:
    def __init__(self, price, curr):
        self.price = price
        self.currency = curr

    def getPrice(self):
        print(self.currency + " "+ str(self.price))
```

```python
class Product:
    def __init__(self, name, brand, price):
        self.name = name
        self.brand = brand
        self.price = price

    def getDetails(iPhone):
        print(iPhone.name)
        print(iPhone.brand)
        print(iPhone.price.getPrice())
```

```python
price = Price(10000,"AUD")
```

```python
iPhone = Product("iphone", "apple", price)
```

```python
print(iPhone.name)
print(iPhone.brand)
print(iPhone.price)
```

```python
print(iPhone.price.price)
```

```python
print(iPhone.price.currency)
```
