## map()

```python
l = [1,2,3,5,6]
sq = [i*i for i in l]
print(sq)
```

```python
def sq(l):
    for i in l:
        print(i**2) 
```

```python
l = [1,2,3,5,6]
sq(l)
```

```python
## map(lambda x:x**2, l)
```

```python
a= list(map(lambda x:x**2, l))
print(a)
```

```python
## odd and even
l = [1,2,3,5,6,7,8,9]
a = list(map(lambda x:'even' if x%2==0 else 'odd', l))
```

```python
a
```

## filter()

```python
l = [1,2,3,5,6,7,8,9]
```

##### number greater than three

```python
list(filter(lambda x:x>3, l))
```

```python
fruits = ['apple','banana','cherry']
# find the name of the fruit with 'a' 
list(filter(lambda x: x.startswith('a'), fruits))
```

## reduce()

```python
import functools
```

```python
l = [1,2,3,5,6,7,8,9]
```

```python
functools.reduce(lambda x,y: x+y, l)
```

```python
l = [1,2,3,5,6,7,8,9]
# find the max value
functools.reduce(lambda x,y:x if x>y else y,l)
```

```python
l = [1,2,3,5,6,7,8,9]
# find the min value
functools.reduce(lambda x,y:x if x<y else y,l)
```
