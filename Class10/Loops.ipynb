### Continue 

It will go straight to the beginning of the loop beside the while condition

```python
start = -2
end = 2

while ( start < end ):
    if (start==0):
        start+=1 
        continue # this will ignore the error and with the increment of the start the loop will start to work again
    
    print(100/start)
    start +=1
```

```python
start = 0
end = 6

while (start<end):
    print(start+1)
    start+=1
    if(start%2==0):
        print('Now I am even and I am inside if condition')
        continue                        # this continue is not the end of the loop
    print('I am odd I will go anyway bcuz I am in a loop')
```

```python
start = 0
end = 4

while (start<end):
    start+=1
    if(start%2==0):
        continue   # the continue will take you in the condition beside the while loop that is at the start
    print(start+1)
```

### Break

Ideally, loop only ends when its condition becomes false

BUT BUT BUT, we can also end the loop comes out of the iteration by using "break" keyword 

```python
start = 1
end = 4

while(start < end):
    if (start %2==0):
        break      # for this break statement the whole thing will go outside the loop
    print(start)
    start+=1
print('I am outside the loop')
```

```python
start = -2
end = 2

while ( start < end ):
    if (start==0):
        start+=1 
        break # it will immidiately break the loop and come outside
    
    print(100/start)
    start +=1
print('I am out side the loop and see the value of start is ', start)
```

```python
start = [-2,-1,0,1]

for number in start:
    print('number is  '+str(number))

print("I am outside the loop")
```

```python
start = [-2,-1,0,1]

for number in start:
    print('number is  '+str(number))
    number = number+5 # it will change nothing in flow of for loop going in the sequence
    

print("I am outside the loop")
```

```python
start = [-2,-1,0,1]

for number in start:
    print('number is  '+str(number))
    number = number+5 # it will change nothing in flow of for loop going in the sequence
    print('Now the number is ' + str(number))

print("I am outside the loop")
```

```python
name = 'python'
for num in name:
    print("alphabet = "+ str(num))
```

```python
name = 'python'
for num in name[::2]:
    print("alphabet = "+ str(num))
```

```python
name = 'python'
alt = 1
for num in name:
    alt = alt +1
    if(alt%2==0):
        print("alphabet = "+ str(num)) # it is following the sequence
```

```python
name = 'python'
alt = 1
for num in name:
    alt = alt +1
    if(alt%2!=0):
        continue
    print("alphabet = "+ str(num)) # it is following the sequence
```
