### WAP to check if the key already exists in a dictionary or not
d =  {1: 10, 2: 20, 3: 30, 4: 40, 5: 50, 6: 60}

it take a value form the user and then check in the dictionary

```python
d = {1: 10, 2: 20, 3: 30, 4: 40, 5: 50, 6: 60}
user = int(input("Please enter an integer to check the value of the key"))

flag = False

for key in d.keys():
    if user == key :
        flag = True
        break

if flag:
    print('yes')
else:
    print('no')
```

```python
d = {1: 10, 2: 20, 3: 30, 4: 40, 5: 50, 6: 60}
user = int(input("Please enter an integer to check the value of the key"))

if user in d:
    print('yes')
else:
    print('no')
```

### WAP find the sum of the values

d = {'data1':100,'data2':-54,'data3':247}

```python
d = {'data1':100,'data2':-54,'data3':247}

s = 0

for val in d.values():
    s += val
print(s)
```

# zip()  

it is used to put the key and value in dictionary


## example


```python
keys = ['a','b','c','d','e']
values = [61,62,63,64,65]

asci = {k:v for (k,v) in zip(keys,values)}

print(asci)
```

```python
a = list(zip(keys,values))
a
```

```python
for i in a:
    print(i)
```

```python
for i,j in a:
    print(i, j)
```
