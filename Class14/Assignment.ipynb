## WAP to count the number of item in a dictionary is list

dict = {'Alex' : ['sub1','sub2','sub3'], 'David' : ['sub1','sub2']}

```python
dic = {'Alex' : ['sub1','sub2','sub3'], 'David' : ['sub1','sub2']}
count = 0
for val in dic.values():
    count += len(val)
print(count)
```

dict = {'Alex' : ['sub1','sub2','sub3'], 'David' : ['sub1','sub2'], 'Alice' : 'sub1'}

```python
dic = {'Alex' : ['sub1','sub2','sub3'], 'David' : ['sub1','sub2'], 'Alice' : 'sub1'}

count = 0

for val in dic.values():
    if isinstance(val,list): # isinstance() is used to check the class of the object
        count += len(val)
print(count)
```
