### WaP a program to accept a name for the user and output it back by removing all in vowels in it

```python
name = input('Enter your name')

ans = ""

for alpha in name:
    if alpha in 'aeiou' or alpha in 'AEIOU':
        continue
    ans = ans+alpha
    
print(ans)
```

#### Write a program to accept an integer from the user and display the sum of all the numbers from 1 to that number.

```python
num = int(input('Enter a num'))
result = 0

for i in range(1, num+1):
    print(i)
    result += i
    
print(f'Sum is {result}')
```

#### Take an input form the user and check if its a palindrome

```python
s = input('Please enter a name')

start = 0
end = len(s)-1
is_palindrome = True

while (start < end):
    if(s[start] != s[end]):
        is_palindrome = False
        break
    else:
        start += 1
        end -= 1
if (is_palindrome == True):
    print('It is palindrome')
else:
     print('It is not a palindrome')   
```

```python
s = input('Please enter a name: ')

is_palindrome = True
n = len(s)

for i in range(n // 2):
    if s[i] != s[n - i - 1]:
        is_palindrome = False
        break

if is_palindrome:
    print('It is a palindrome')
else:
    print('It is not a palindrome')
```

```python
n = int(input('Please enter a number'))

for i in range(n):      # handles column
    print('*', end="")
    for j in range (n-1): # handles row
        print('*',end="")
    print()
    
```

```python
n = int(input('Please enter a number'))

for i in range(n):       # It handles number of rows
    for j in range (n):  # It handles inner works in Row
        print('*',end="")
    print()
```
