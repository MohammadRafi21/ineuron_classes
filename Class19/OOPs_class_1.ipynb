```python
class atm_machine:


  #constructor (special function)  - superpower

  def __init__(self): # this is called constructor
  # all the Data properties of a class will be inside this constructor
    self.pin = ""     # only atm_machine class can access this data pin and balance
    self.balance = 0  # that is why we are using "self." before pin and balance
    self.menu()

    # the funtion inside the class

  def menu(self):
    user_input = input("""
    Hello, how would you like to proceed?
    1. Enter 1 to create pin
    2. Enter 2 to change pin
    3. Enter 3 to check balance
    4. Enter 4 to withdraw
    5. Enter anything to exit
    """
    )

    if user_input == "1":

      # creat a pin
      self.create_pin()

    elif user_input == "2":

      # change pin
      self.change_pin()

    elif user_input == "3":

      # check balance
      self.check_balance()

    elif user_input == "4":

      # withdraw
      self.withdraw()

    else:
      exit()

  def create_pin(self):

    user_pin = input("Please Enter your pin")
    self.pin = user_pin

    user_balance = int(input("Please Enter your balance"))
    self.balance = user_balance

    print("PIN has been created sucessfully")
    self.menu()

  def change_pin(self):
    old_pin = input("Please enter the Old PIN")
    if old_pin == self.pin:
      new_pin = input("Please enter the New PIN")
      self.pin = new_pin
      print("New PIN has been created sucessfully")
      self.menu()
    else:
      print("Old PIN is wrong")
      self.menu()

  def check_balance(self):
    user_pin = input("Enter your pin")
    if user_pin == self.pin:
      print(f"your balance is {self.balance}")
    else:
      print("You entered wrong PIN")
    self.menu()


  def withdraw(self):
    user_pin = input("Enter your pin")
    if user_pin == self.pin:
      # allow to withdraw
      withdraw_amount = int(input("Enter the amount you want to withdraw"))
      if withdraw_amount <= self.balance:
        self.balance = self.balance - withdraw_amount
        print(f"You have withdrawn {withdraw_amount}")
        print(f"Your remaining balance is {self.balance}")
      else:
        print("You have insufficient balance")
    else:
      print("Ypu entered wrong PIN")
    self.menu()

```

```python
obj = atm_machine()
# no need to call the menu() fuction seperately because the constructor is doing that work "self.menu()"
```
