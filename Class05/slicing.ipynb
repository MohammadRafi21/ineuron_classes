# Slicing

##### when you try to excess a single element from a sequence
##### [] is used for slicing
##### [start : end]

```python
# indexing

s1= 'PYTHON'
```

```python
print(s1[0])
print(s1[1])
print(s1[2])
print(s1[3])
print(s1[4])
print(s1[5])
```

```python
print(s1[-1]) # -1 + len(s1) = 5
print(s1[-2])
print(s1[-3])
print(s1[-4])
print(s1[-5])
print(s1[-6])
```

### how slicing is working

##### s1[start : end]  the slicing will start at the start index and it will end 1 step before the end index

```python
s1[0:4] # it will take the first but it will stop 1 step before
```

```python
s1[0:len(s1)]
```

```python
s1[1:4]
```

```python
s1[3:5]
```

```python
s1[3:] # if there is nothing in the end index python will by default will take you to the end index
```

```python
s1[:4]
```

```python
s1[:]
```

```python
s1[0:1]
```

```python
s1[0:0]
```
