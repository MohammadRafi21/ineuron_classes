# It you use the input() fuction in python the input that it takes form the user it take it as a String so for calculation perpose you have to convert the input data on your own

```python
a = input("Please enter a Number")
b = input("Please enter another Number for substraction")
c = int(a) * int(b)
print(type(c))
print("the answer is "+ str(c))
```

```python
# Another way
a = int(input("Please enter a Number"))
b = int(input("Please enter another Number for substraction"))
c = a * b
print(type(c))
print("the answer is "+ str(c))
```
