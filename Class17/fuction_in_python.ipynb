## args & kwrgs

args and kwrgs are special python keywords tht are used to pass the variable lenght to a function 

## args
allows us to pass a variable of non- keyword argument to a function



```python
def adds(a, b):
    return a+b
```

```python
adds(2, 3)
```

```python
def adds(a, b, c):
    return a+b+c
```

```python
adds(2, 3, 5)
```

```python
def adds(*args):
    output = 0

    for i in args:
        output += i
    print(args)

    return output
```

```python
adds(1,5,6,7,8,9,3)
```

```python
def adds(*summation): # here "*" works as a args
    output = 0

    for i in summation:
        output += i
    print(summation)

    return output

adds(1,5,6,7,8,9,3)
```

## kwargs
kwargs allows us to pass any number of keyword arguments

```python
def display_info(**kwargs): ## this kwargs create a dictonary
    print(kwargs)
```

```python
display_info(india='delhi',srilanka='colombo',nepal='kathmandu',pakistan='islamabad')
```

```python
def display_info(**kwargs):
    for k,v in kwargs.items():
        print(k,'->',v)
    print(kwargs)
```

```python
display_info(india='delhi',srilanka='colombo',nepal='kathmandu',pakistan='islamabad')
```

## higher order function

```python
def square(x):
    return x**2

l= [2,3,4,5,6,7]

for i in l:
    print(square(i))
```

```python
l= [2,3,4,5,6,7]

def square(x):  # this square functioncan be written in lambda
    return x**2

def transform(f, l):
    output = []

    for i in l:
        output.append(f(i))

    return output
result = transform(square, l)
print(result)

```

```python
l= [2,3,4,5,6,7]

def transform(f, l):
    output = []

    for i in l:
        output.append(f(i))

    return output
result = transform(lambda x:x**2,l)
print(result)
```
