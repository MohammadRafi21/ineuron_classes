#### Wap where input to take input a char form the user and show below

1. A-Z Capital letter
2. a-z small letter
3. 0-9 number
4. special character

```python
char = input("Please enter the Character")

asciiValue = ord(char)

if (asciiValue >= 97 and asciiValue <= (97+26-1)):
    print('smaller letter')
elif (asciiValue >= 65 and asciiValue <= (65+26-1)):
    print('Capital letter')
elif (asciiValue >= 48 and asciiValue <= (48+9)):
    print('it is a digit')
else:
    print('It s a special char')
```

Wap to input a year form user and check if it is a leap year

```python
year= int(input('Please enter a year'))

if (year%4==0):
    if(year%100==0):
        if(year%400==0):
            print(f"{year} is a leap year")
        else:
            print(f"{year} is NOT a leap year")
    else:
        print(f"{year} is NOT a leap year")        
else:
    print(f"{year} is NOT a leap year")
```
