```python
def prime_factor(n,divisor):
    l = []
    while n !=1:
        if n % divisor == 0:
            l.append(divisor)
            n = int(n/divisor)
        else:
            divisor +=1
    return l



divisor = 2

x = int(input())
y = int(input())

factors1 = prime_factor(x,divisor)

factors2 = prime_factor(y,divisor)

factors1.sort()
factors2.sort()

print(factors1)
print(factors2)

```

```python
i = 0
j = 0
gcd = 1

while i<len(factors1) and j<len(factors2):
    if (factors1[i] == factors2[j]):
        gcd *= factors1[i]
        i += 1
        j += 1
    elif (factors1[i] > factors2 [j]):
        j +=1
    else:
        i +=1

print(gcd)
```
