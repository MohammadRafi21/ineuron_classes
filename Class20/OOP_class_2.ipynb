```python
class Fraction:
    def __init__(self, num = 1, den = 1):
      
      self.num = num # instance variable
      self.den = den
      
      if(den == 0):
          print("Denominator cannot be 0")
          self.num = num
          self.den = 1
          
    def __str__(self):
        return "{}/{}".format(self.num,self.den)
  
    def __add__(self,fraction2):
       self.num = self.num*fraction2.den + self.den*fraction2.num
       self.den = self.den * fraction2.den
       ans = Fraction(self.num,self.den)
       ans.simplify()
       return ans
       #return "{}/{}".format(self.num,self.den)

    def myaddFuction(self,fraction2):
       self.num = self.num*fraction2.den + self.den*fraction2.num
       self.den = self.den * fraction2.den
       ans = Fraction(self.num,self.den)
       ans.simplify()
       return ans
       
    def __sub__(self, fraction2):
       self.num = self.num*fraction2.den - self.den*fraction2.num
       self.den = self.den * fraction2.den
       ans = Fraction(self.num,self.den)
       ans.simplify()
       return ans
       #return "{}/{}".format(self.num,self.den)
    def __mul__(self, fraction2):
       self.num = self.num*fraction2.num
       self.den = self.den * fraction2.den
       ans = Fraction(self.num,self.den)
       ans.simplify()
       return ans
       #return "{}/{}".format(self.num,self.den)
    def __truediv__(self, fraction2):
       self.num = self.num*fraction2.den
       self.den = self.den * fraction2.num
       ans = Fraction(self.num,self.den)
       ans.simplify()
       return ans
       #return "{}/{}".format(self.num,self.den)

   #  def __pow__(firstOperand, second_operand):
   #     self.num = self.num
   #     return "{}/{}".format(self.num,self.den)
    
    def simplify(self):
       
        ## get the GCD of numerator and denominator and return the ans after dividing by it
       ## make sure that all operation above return the ans after simplifying

       divisor = 2
       factors1 = self.prime_factor(self.num,divisor)
       factors2 = self.prime_factor(self.den,divisor)
       factors1.sort()
       factors2.sort()
       gcd = self.common_gcd(factors1,factors2)

       self.num = int(self.num/gcd)
       self.den = int(self.den/gcd)

       

   
      
    def prime_factor(self, n, divisor):
       
       l = []

       while n !=1:
         if n % divisor == 0:
            l.append(divisor)
            n = int(n/divisor)
         else:
            divisor +=1
            
       return l
    def common_gcd(self,l1,l2):
       i = 0
       j = 0
       gcd_c = 1
       while i<len(l1) and j<len(l2):
          if (l1[i] == l2[j]):
             gcd_c *= l1[i]
             i += 1
             j += 1
          elif (l1[i] > l2 [j]):
             j +=1
          else:
             i +=1
       return gcd_c

       


 
```

```python
class Mentor():

    def __init__(self, name, number):
        self.name = name
        #self.number = number
        self.__number = number # this "__" makes it private

    def __str__(self):
        print(self.name)
        print(self.__number)
        return ""
```

```python
mentor1 = Mentor("rafi","467465")
```

```python
print(mentor1)
```

```python
print(mentor1.name) # can access the name because it is public 
```

```python
print(mentor1.__number) # cannot access the number because it is private
```
