## we cannot define multuple constructor in python, the latest one overrides the old one

```python
class student:

  def __init__(self):
    self.__math = 0
    self.science = 0


  def __init__(self, math, science):
    self.math = math
    self.science = science
```

```python
s1 = student()
s2 = student(10,20)
```

## in the above code the latest "__init__" constructor is overriding the previous "__init__" constructor

```python
s2 = student(10,20) # now it is working
```

```python
class Fraction:
    def __init__(self, num = 1, den = 1):
      
      self.num = num
      self.den = den
      
      if(den == 0):
          print("Denominator cannot be 0")
          self.num = num
          self.den = 1
          
    def __str__(self):
        return "{}/{}".format(self.num,self.den)
  
    def __add__(self,fraction2):
       self.num = self.num*fraction2.den + self.den*fraction2.num
       self.den = self.den * fraction2.den
       ans = Fraction(self.num,self.den)
       return ans
       #return "{}/{}".format(self.num,self.den)
       
    def __sub__(self, fraction2):
       self.num = self.num*fraction2.den - self.den*fraction2.num
       self.den = self.den * fraction2.den
       return "{}/{}".format(self.num,self.den)
    def __mul__(self, fraction2):
       self.num = self.num*fraction2.num
       self.den = self.den * fraction2.den
       return "{}/{}".format(self.num,self.den)
    def __truediv__(self, fraction2):
       self.num = self.num*fraction2.den
       self.den = self.den * fraction2.num
       return "{}/{}".format(self.num,self.den)
    
    def simplify(self):
       
        ## get the GCD of numerator and denominator and return the ans after dividing by it
       ## make sure that all operation above return the ans after simplifying

       divisor = 2
       factors1 = self.prime_factor(self.num,divisor)
       factors2 = self.prime_factor(self.den,divisor)
       factors1.sort()
       factors2.sort()
       gcd = self.common_gcd(factors1,factors2)

       self.num = int(self.num/gcd)
       self.den = int(self.den/gcd)

       

   
      
    def prime_factor(self, n, divisor):
       
       l = []

       while n !=1:
         if n % divisor == 0:
            l.append(divisor)
            n = int(n/divisor)
         else:
            divisor +=1
            
       return l
    def common_gcd(self,l1,l2):
       i = 0
       j = 0
       gcd_c = 1
       while i<len(l1) and j<len(l2):
          if (l1[i] == l2[j]):
             gcd_c *= l1[i]
             i += 1
             j += 1
          elif (l1[i] > l2 [j]):
             j +=1
          else:
             i +=1
       return gcd_c

       


 
```
