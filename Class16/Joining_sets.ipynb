# union() & update()

s.union(t)  s | t   new set with elements from both s and t

s.update(t) s |= t  return set s with elements added from t

```python
city_name1 = {'Tokyo','Madrid','Berlin','Delhi'}
city_name2 = {'Tokyo','Seoul','kabul','Madrid'}

city_name3 = city_name1.union(city_name2)
city_name4 = city_name1.intersection(city_name2)

print(city_name3)
print(city_name4)
```

```python
num1 = {1,2,3,4}
num2 = {3,4,5,6}

num3 = num1.union(num2)
```

```python
num3
```

```python
num4 = num1.intersection(num2)
num4
```

```python
city_name1 = {'Tokyo','Madrid','Berlin','Delhi'}
city_name2 = {'Tokyo','Seoul','kabul','Madrid'}

city_name1.update(city_name2) # same as the union fuction
print(city_name1)
```

# intersection() & intersection_update()

s.intersection_update(t)    s &= t  return set s keeping only elements also found in t

s.intersection(t)   s & t   new set with elements common to s and t

```python
city_name1 = {'Tokyo','Madrid','Berlin','Delhi'}
city_name2 = {'Tokyo','Seoul','kabul','Madrid'}

city_name3 = city_name1.intersection(city_name2)
print(city_name3)
```

```python
num1 = {1,2,3,4}
num2 = {3,4,5,6}

num4 = num1.intersection(num2)
num4
```

```python
city_name1 = {'Tokyo','Madrid','Berlin','Delhi'}
city_name2 = {'Tokyo','Seoul','kabul','Madrid'}

city_name1.intersection_update(city_name2)
city_name1
```

### isdisjoint()


the isdisjoint() method checks if items of given set are present in another set. the method returns "False" if items are present else it returns "True"

```python
city_name1 = {'Tokyo','Madrid','Berlin','Delhi'}
city_name2 = {'Tokyo','Seoul','kabul','Madrid'}


city_name1.isdisjoint(city_name2)
```

```python
city_name1 = {'Tokyo','Madrid','Berlin','Delhi'}
city_name2 = {'Dhaka','Seoul','kabul','Barcelona'}


city_name1.isdisjoint(city_name2)
```

## pop() for the list

The pop() method in Python, when used on a list, removes and returns an element. 
Without an argument: pop() removes and returns the last element of the list.
With an integer argument (index): pop(index) removes and returns the element at the specified index. For example, pop(0) would remove the first element, pop(1) the second, and so on.


## pop() for the set

In Python, the pop() method for sets removes and returns an arbitrary / random element from the set.



#### If you want to take the last item form a set you have to convert it to a list BUT when the set is converted to a list it will reoreder it self by its own

```python
city_name1 = {'Tokyo','Madrid','Berlin','Delhi'}
city_name1 = list(city_name1)
print(city_name1)
print(city_name1.pop())
city_name1 = set(city_name1)
print(city_name1)
```
