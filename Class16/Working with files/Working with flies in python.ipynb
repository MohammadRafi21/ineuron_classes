| Mode | Can Read? | Can Write? | Behavior |
|------|------------|------------|-----------|
| `'r'` | ‚úÖ | ‚ùå | Read-only |
| `'w'` | ‚ùå | ‚úÖ | Overwrites file |
| `'a'` | ‚ùå | ‚úÖ | Appends to file |
| `'r+'` | ‚úÖ | ‚úÖ | Read & write (no truncate) |
| `'w+'` | ‚úÖ | ‚úÖ | Read & write (overwrites file) |
| `'a+'` | ‚úÖ | ‚úÖ | Read & append (writes at end) |

---

#### üí° Tips:
- **`'r'`** ‚Üí File must exist, only reading allowed.  
- **`'w'`** ‚Üí Creates or overwrites the file.  
- **`'a'`** ‚Üí Adds new data at the end of file.  
- **`'r+'`** ‚Üí Reads and writes without deleting existing data.  
- **`'w+'`** ‚Üí Reads and writes but clears old content first.  
- **`'a+'`** ‚Üí Reads and writes but only appends new data.
