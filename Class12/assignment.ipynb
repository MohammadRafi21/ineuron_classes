## answer to question no 1

```python
lis1 = [i for i in range (20)]
s = 0
for i in lis1:
    s +=lis1[i]
print(s)
```

#### the following code is better because it prevent index error and here i in the for loop for the finding the sum act as a value inside the list it doesn't indicate the indexes of the list

```python
lis1 = [i for i in range (20)]
s = 0
for i in lis1:
    s +=i
print(s)
    
```

## answer to the question no 2

```python
lis2 = [1,8,9,5,6,64,7,10,4,2,15,3]
largest = lis2[0]
for i in lis2:
    if (i>largest):
        largest = i
print(largest)
```

## answer to the question no 3

```python
lis1 = ['banana', 'mango','orange','peach','water melon','grape']

lis2 = []

for i in lis1:
    lis2.append(i)
print(lis2)
```

```python
## other ways

lis3 = lis1[:] ## using the slicing technique
print(lis3)

```

```python
lis4 = list(lis1) ## using the list constructor
print(lis4)
```

```python
lis5 = lis1.copy() ## using the copy method
print(lis5)
```

```python
import copy

lis6 = copy.deepcopy(lis1) ## using the deepcopy method
print(lis6)
```

# answer to the question no 4

```python
lis8 = [1,2,3,5,9,8,6,4,10,12,13,14,20,21,25,26]

a = int(input('Please enter a number'))

flag = 0

for i in lis8:
    if a==i:
        flag = 1
        break
if flag==1:
    print('Found the number')
else:
    print('not found')


```

#### shorter version

```python
lis8 = [1, 2, 3, 5, 9, 8, 6, 4, 10, 12, 13, 14, 20, 21, 25, 26]

a = int(input("Please enter a number: "))

if a in lis8:
    print("Found the number")
else:
    print("Not found")
```

```python
fruits = ['Banana', 'Mango', 'Orange', 'Peach', 'Water Melon', 'Grape']

user = input('Please enter a fruit name').lower()

fruits_lower = [i.lower() for i in fruits]

print(fruits_lower)

if user in fruits_lower:
    print('Found')
else:
    print('not found')
```
