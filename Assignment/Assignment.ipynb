#### WAP to accept a character from the user and check whether it is a capital letter or small letter. Assume user will input only alphabets

```python
while True:
    char = input("Enter a alphabet")
    asci = ord(char)
    if (asci >=65 and asci <= 90  ):
        print('Capital Letter')
        break
    elif (asci >=97 and asci <= 122):
        print('Small letters')
        break
    else:
        print('Not a Alphabet Enter again')
        continue
```

#### WAP to accept a year from the user and check whether it is a leap year or not.

#### You have to develop a number guessing game.
#### Following are the steps needed:
#### ● At first , the program will generate a random integer secretly between 1 to 50.
#### ● Now it will ask the user to guess that number .
#### ● If the user guessed it correctly then the program prints “Congratulations! You guessed it right” .
#### ● But if the number guessed by the user is larger than the secret number then program should print “Number too large” and , if the number guessed by the user is smaller than the secret number then the program should print “Number too small”
#### ● This should continue until the user guesses the number correctly or quits .
#### ● If the user wants to quit in between he will have to type 0 or negative number

```python
import random

ran_num = random.randint(1,50)

# guess = int(input("Please Guess a integer number between 1 and 50 to win"))

while True:
    guess = int(input("Enter your guess (or type 0 or a negative number to quit): "))
    if (guess <= 0):
        print('Thank you for playing! The secret number was ', ran_num)
        break
    elif (guess == ran_num):
        print('Congratulations! You guessed it right')
        break
    elif (guess > ran_num):
        print('Number too large')
    elif (guess < ran_num):
        print('Number too small')
      
```

#### Write a program to accept an integer from the user and display the sum of all the numbers from 1 to that number.

```python
user_num = int(input('Please enter a Number'))
store = []
x = 0

while (x < user_num):
    x += 1
    store.append(x)
add = 0
for num in store:
    add += num
print(store)
print(add)
    
```

```python
# Accept input from the user
user_num = int(input("Please enter a number: "))

# Initialize variables
x = 1
total_sum = 0

# Use a while loop to calculate the sum
while x <= user_num:
    total_sum += x
    x += 1

# Display the result
print(f"The sum of all numbers from 1 to {user_num} is {total_sum}.")
```

### Write a program to accept an integer from the user and calculate its factorial

```python
user_num = int(input('Please enter a Number'))
store = []
x = 0

while (x < user_num):
    x += 1
    store.append(x)
mul = 1
for num in store:
    mul = mul*num
print(store)
print(mul)
```

```python
# Accept an integer from the user
user_num = int(input('Please enter a number: '))

# Initialize variables
factorial = 1
x = 1

# Calculate factorial using a while loop
while (x <= user_num):
    factorial *= x
    x += 1

# Display the result
print(f"The factorial of {user_num} is {factorial}.")
```
