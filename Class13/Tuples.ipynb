### to add item in tuple we have to do it indirectly 

1. first convert it to a list
2. add item in the list
3. then convert it to tuple again

```python
tup = (1,2,3,4,5)
```

```python
print(tup)
```

```python
temp = list(tup)
print(temp)
```

```python
temp.append(7)
print(temp)
```

```python
tup = tuple(temp)
print(tup)
```
