## remove()

remove the first items with the specified value it doesn't take the index number it take the element itself 

```python
fruits = ['apple','banana', 'mango','cherry','banana','Banana']
```

```python
fruits.remove('banana')
```

```python
print(fruits) ## removed the 1st banana form the list
```

```python
temp = []
for i in fruits:
    if i.lower() == 'banana':
        continue
    else:
        temp.append(i)
```

```python
print(temp)
```

```python
temp = []
for i in fruits:
    if i.lower() == 'banana':
        pass
    else:
        temp.append(i)
```

```python
print(temp)
```

### this is side quest that is to understand the difference between the pass and the continue statement in above code you can see the continue and the pass statement is doing the same thing but which is not actually the same thing in the codes below you can see the differences so "Pass" statement actually Does nothing â€” placeholder but "continue" statement 	Skips the rest of the loop for current item

```python
fruits = ['apple','banana', 'mango','cherry','banana','Banana']
temp = []
for i in fruits:
    if i.lower() == 'banana':
        continue
    temp.append(i)
print(temp)
```

```python
fruits = ['apple','banana', 'mango','cherry','banana','Banana']
temp = []
for i in fruits:
    if i.lower() == 'banana':
        pass
    temp.append(i)
print(temp)
```
