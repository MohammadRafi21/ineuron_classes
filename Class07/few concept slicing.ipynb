# few concept of slicing

## using +ve and -ve inside slicing

```python
s= 'PYTHON'
```

```python
s[-4:4] # so it becomes s[2:4] or s[-4:-2]
```

## See the notes for clear view on it

```python
s[-4:0:1] # we are trying to go in opposite direction which is not possible (s[-4:-6:1] or s[2:0:1] concept)
```

```python
s[-4:0:-1] # the concept is s[-4:-6:-1] or s[2:0:-1] answer will be "TY"
```

## Default Value for -ve stride 


```python
s[::-1] # it is moving backward
```

```python
s[:3:-1] # see the note
```

```python
s[:4:-1] # see the note
```

```python
s[2::-1] # see the note
```

```python
s[:-4:-1] # see the note
```

```python
s[-3::-1] # see the note
```

# Nested Slicing

```python
s = 'PYTHON IS GREAT'
```

```python
# form the string lets say just take the python
s[0:6:1]
```

```python
s2 = s[0:6]
```

```python
s2[0:3] # it is taking form the string  'PYTHON'
```

### now this is the advance slicing here we can see the thing we did above is done in one line of code

```python
s[0:6][0:3]
```
