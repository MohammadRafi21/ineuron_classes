# Logical Operators

### Logical Operators, they help to combine 2 or more conditions and performs the total operations
### AND (True when all are TRUE)
### OR (True when either is TRUE)
### NOT (Changers to True when False ) or negate your result

### and

```python
5 > 4 and 3 > 2
```

```python
'a' > 'A' and 5>3 and True > False and 6.34 > 6.33
```

```python
'a' > 'A' and 5>3 and True > False and 6.34 < 6.33
```

## or

```python
5>3 or 'a' > 'x'
```

## not

```python
not(5>3 or 'a' > 'x')
```

```python
not (True)
```

```python
not(False)
```

```python
not (5 > 4 and 4 > 6)
```
